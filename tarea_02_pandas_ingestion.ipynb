## Parte 2 — Selección y filtrado

**P4.** Selecciona solo las columnas `sepal_length` y `species` e imprime las primeras 5 filas.

```python
# Tu respuesta aquí

columnas = iris[['sepal_length', 'species']]
print(columnas.head(5))
```

**P5.** Filtra el DataFrame para quedarte solo con las flores de la especie `virginica`.  
¿Cuántas filas tiene ese subconjunto?

```python
# Tu respuesta aquí

# Filtramos donde la especie sea exactamente 'virginica'
virginica_df = iris[iris['species'] == 'virginica']

# Mostramos el resultado
print(virginica_df.head())

# Verificamos cuántas filas encontramos
print(f"\nTotal de registros de virginica: {len(virginica_df)}")
```

**P6.** Filtra las filas donde `petal_length` sea mayor a 5.0. Imprime cuántas filas cumplen esa condición.

```python
# Tu respuesta aquí

# Filtramos las filas donde petal_length es mayor a 5.0
petal_grande = iris[iris['petal_length'] > 5.0]

# Imprimimos cuántas filas cumplen la condición
print(f"Número de filas con petal_length > 5.0: {len(petal_grande)}")
```

## Parte 3 — Guardar datos

**P7.** Guarda el subconjunto de `virginica` (de P5) en un archivo CSV llamado `virginica.csv` en la carpeta `output/`.  
No incluyas el índice en el archivo guardado.

```python
import os
os.makedirs("output", exist_ok=True)

# Tu respuesta aquí

if not os.path.exists('output'):
    os.makedirs('output')

virginica_df = iris[iris['species'] == 'virginica']

virginica_df.to_csv('output/virginica.csv', index=False)

print("Archivo guardado exitosamente en output/virginica.csv")
```
