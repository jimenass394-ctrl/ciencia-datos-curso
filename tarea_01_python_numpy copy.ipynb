**P6.** Crea los siguientes arreglos e imprime su `shape` y `dtype`:

```python
# Tu respuesta aquí

# 1. Arreglo con valores específicos
a = np.array([1.5, 2.5, 3.5, 4.5])

# 2. Matriz de ceros de 3 filas x 4 columnas
b = np.zeros((3, 4))

# 3. Arreglo con 6 valores igualmente espaciados entre 0 y 1
c = np.linspace(0, 1, 6)

# Impresión de resultados
print(f"Arreglo A: {a}")
print(f"Shape: {a.shape}, Dtype: {a.dtype}")
print("-" * 30)

print(f"Arreglo B (Ceros):\n{b}")
print(f"Shape: {b.shape}, Dtype: {b.dtype}")
print("-" * 30)

print(f"Arreglo C (Linspace): {c}")
print(f"Shape: {c.shape}, Dtype: {c.dtype}")
```

**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)

# Tu respuesta aquí

# 1. Crear el arreglo del 1 al 12
a = np.arange(1, 13)

# 2. Convertirlo en una matriz de 3 filas x 4 columnas
matriz = a.reshape(3, 4)

# 3. Obtener estadísticas
maximo = matriz.max()
minimo = matriz.min()
posicion_maximo = matriz.argmax()

# Impresión de resultados
print("Matriz resultante (3x4):")
print(matriz)
print(f"\nValor máximo: {maximo}")
print(f"Valor mínimo: {minimo}")
print(f"Posición ndice valor máximo: {posicion_maximo}")
```

**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).

```python
# Tu respuesta aquí

# 1. Crear dos arreglos de 5 elementos con valores entre 1 y 20
# Nota: el límite superior en randint es exclusivo, por eso usamos 21
a = np.random.randint(1, 21, size=5)
b = np.random.randint(1, 21, size=5)

# 2. Realizar las operaciones
suma = a + b
resta = a - b
multiplicacion = a * b

# 3. Imprimir los resultados
print(f"Arreglo A: {a}")
print(f"Arreglo B: {b}")
print("-" * 30)
print(f"Suma:           {suma}")
print(f"Resta:          {resta}")
print(f"Multiplicación: {multiplicacion}")
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Tu respuesta aquí

# 1. Crear el arreglo de notas
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# 2. Aplicar indexación booleana
# notas >= 6 crea una "máscara" de valores True/False
notas_aprobadas = notas[notas >= 6]

# 3. Imprimir el resultado
print(f"Todas las notas: {notas}")
print(f"Notas mayores o iguales a 6: {notas_aprobadas}")
```
